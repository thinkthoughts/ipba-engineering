## Load the repository in Colab or local Jupyter

```python
from pathlib import Path
import os
import subprocess
import sys

REPOSITORY_URL = "https://github.com/thinkthoughts/information-bundling-auctions.git"
PACKAGE_NAME = "information_bundling_auctions"
CURRENT = Path.cwd().resolve()

ROOT = next(
    (candidate for candidate in (CURRENT, CURRENT.parent)
     if (candidate / "src" / PACKAGE_NAME).exists()),
    None,
)

if ROOT is None:
    ROOT = Path("/content/information-bundling-auctions")
    if not ROOT.exists():
        subprocess.run(
            ["git", "clone", REPOSITORY_URL, str(ROOT)],
            check=True,
        )

SRC = ROOT / "src"
PACKAGE_DIR = SRC / PACKAGE_NAME

if not PACKAGE_DIR.exists():
    raise RuntimeError(f"Expected Python package at {PACKAGE_DIR}")

if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

os.chdir(ROOT)

print("Repository root:", ROOT)
print("Python source:", SRC)
print("Working directory:", Path.cwd())
```

```python
from IPython.display import Markdown, display

display(Markdown(
    f"# Engineering Context\n\n"
    f"**Repository:** `{CONTEXT.repository}`  \n"
    f"**Repository description:** *{CONTEXT.repository_description}*\n\n"
    f"## {CONTEXT.engineering_statement_title}\n\n"
    f"> **{CONTEXT.engineering_statement}**\n\n"
    "Notebook 00 specifies the repository vocabulary, engineering variable, "
    "notebook sequence, and generated context artifacts used by every later notebook."
))
```

```python
from IPython.display import Markdown, display

grammar_chain = " → ".join(CONTEXT.grammar)
display(Markdown(
    f"## {CONTEXT.grammar_title}\n\n"
    f"**{grammar_chain}**\n\n"
    f"**{CONTEXT.design_principle}**"
))
```

```python
from IPython.display import Markdown, display

lane_lines = []
for label, relationship in zip(
    CONTEXT.lane_labels[:-1], CONTEXT.lane_relationships
):
    lane_lines.append(f"**{label}** {relationship}")

display(Markdown(
    f"## {CONTEXT.repository_variable_title}\n\n"
    f"**{CONTEXT.connected_lane}**\n\n"
    f"{CONTEXT.design_principle}"
))
```

```python
from IPython.display import Markdown, display

sequence_lines = "\n".join(
    f"- {item}" for item in CONTEXT.construction_sequence
)
display(Markdown(
    f"## {CONTEXT.repository_sequence_title}\n\n"
    f"{CONTEXT.repository_sequence_caption}\n\n"
    f"{sequence_lines}"
))
```
